# Task 2

Objective: Create a restaurant recommendation
system based on user preferences.

Steps:
Preprocess the dataset by handling missing
values and encoding categorical variables.
Determine the criteria for restaurant
recommendations (e.g., cuisine preference,
price range).
Implement a content-based filtering
approach where users are recommended
restaurants similar to their preferred criteria.
Test the recommendation system by
providing sample user preferences and
evaluating the quality of recommendations.

```python

import numpy as np
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity


```

```python
data = pd.read_csv("/content/Dataset .csv")

```

```python
data['Cuisines'] = data['Cuisines'].fillna('Unknown')
data['Aggregate rating'] = data['Aggregate rating'].fillna(data['Aggregate rating'].mean())
data['Price range'] = data['Price range'].fillna(data['Price range'].mode()[0])


data['City'] = data['City'].astype('category').cat.codes

```

```python
tfidf = TfidfVectorizer(stop_words='english')
cuisine_matrix = tfidf.fit_transform(data['Cuisines'])


data['features'] = data.apply(lambda x: f"{x['Cuisines']} {x['Price range']} {x['Aggregate rating']}", axis=1)
feature_matrix = tfidf.fit_transform(data['features'])


```

```python
def recommend_restaurants(preferences, top_n=5):
    query = f"{preferences['cuisine']} {preferences['price_range']} {preferences['min_rating']}"
    query_vec = tfidf.transform([query])
    similarity_scores = cosine_similarity(query_vec, feature_matrix)

    data['similarity'] = similarity_scores.flatten()
    filtered_data = data[(data['City'] == preferences['city']) & (data['Aggregate rating'] >= preferences['min_rating'])]

    return filtered_data.sort_values(by='similarity', ascending=False).head(top_n)

```

```python
user_preference = {
    "cuisine": "Japanese",
    "price_range": 3,
    "min_rating": 4.3,
    "city": 0
}


```

```python
recommendations = recommend_restaurants(user_preference)
dFF = pd.DataFrame(recommendations[['Restaurant Name', 'Cuisines', 'Aggregate rating', 'Price range']])
dFF
```
